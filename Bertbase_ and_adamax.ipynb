### Definition of classes. 

```python
# We define out classes for ease of modelling 
class Warehouse_Model(object):
    """Class to save your model architecture and Model Weights correctly"""
    def __init__(self):
        pass
   
    def save(model,file_path):
        """
        Function for saving both the model and weights to your file_path
        Follow this structure ['/file_path/name_you_want_to_save_as']
        """
        SaveModel = tf.saved_model.save(model, file_path)
        
        
        SaveWeights = model.save_weights(file_path + '.h5')
        
        return print(f'Model and Weights saved SUCCESSFULLY in:{file_path}')
    
    def load(file_path):
        
        """Follow this structure ['/file_path/name_of_your_model']"""
        # Load The Model
        LoadModel = tf.saved_model.load(file_path)
        
        return LoadModel
```

```python
# Defining a Class for loading transformers from Hugging face.
class LoadBert(object):
    
    """A class that loads a bert model, tokenizer and a transformer"""
    def __init__(self):
        pass
    def loadModel(modelName):
        """Loading the model name"""
        model = modelName
        return model

    def config (Bert_Configuration, modelName, outputhidden_states:bool):
        """Model configuration"""
        config = Bert_Configuration.from_pretrained(modelName)
        config.outputhidden_states = False
        return config

    def tokenizer(tokenizer, modelName, config):
        """Text tokenizer"""
        tokenizer = tokenizer.from_pretrained(pretrained_model_name_or_path = modelName, config =config)
        return tokenizer

    def transformerModel(BertClass, modelName, config):
        """Hugging face transformer model"""
        transformerModel = BertClass.from_pretrained(modelName, config)
        return transformerModel
```

```python
# Readying the Output
class ready_output(object):
    """ A class to ready output for train and test data"""
    def __init__(self):
        pass
    
    def load_data(filename:str):
        
        """ A function that loads data"""
        output = pd.read_csv(filename, index_col = 0)
        return output

    def get_target(df, target):
        """readying the target output"""
        return to_categorical(df[target])
    
    def get_tokens(tokenizer, df, text_col, max_len:int):
        """A function to get test and train tokens"""
        tokens = tokenizer(text=df[text_col].to_list(),
                        add_special_tokens=True,
                        max_length=max_len,
                        truncation=True,
                        padding=True, 
                        return_tensors='tf',
                        return_token_type_ids = False,
                        return_attention_mask = True,
                        verbose = True,
                               
                        )
        return tokens
```

```python
def log_text(epoch, logs, text_data):
    with text_writer.as_default():
        tf.summary.text("text_data", text_data, step=epoch)
    text_writer.flush()
```

## Loading the cleaned dataset.

```python
# Loading the data
loadData = ready_output
data = loadData.load_data('data/df_ml.csv')


# previewing the data
data.head()
```

```python
#  Dropping rows with missing values

data = data.dropna(subset=['clean_narrative'])
```

```python
# Isolating the relevant columns

data_deep = data[['product','company_response_to_consumer','timely_response',
                    'consumer_disputed','clean_narrative']]
```

```python
# Remove rows, where the label is present only ones (can't be split)
data_deep = data_deep.groupby('product').filter(lambda x : len(x) > 1)
data_deep = data_deep.groupby('company_response_to_consumer').filter(lambda x : len(x) > 1)
data_deep = data_deep.groupby('timely_response').filter(lambda x : len(x) > 1)
data_deep = data_deep.groupby('consumer_disputed').filter(lambda x : len(x) > 1)

# Set model outputs as categorical and save in new label col
data_deep['product_label'] = pd.Categorical(data_deep['product'])
data_deep['responseto_label'] = pd.Categorical(data_deep['company_response_to_consumer'])
data_deep['timely_response_label'] = pd.Categorical(data_deep['timely_response'])
data_deep['disputed_label'] = pd.Categorical(data_deep['consumer_disputed'])

# Transform your output to numeric
data_deep['product'] = data_deep['product_label'].cat.codes
data_deep['company_response_to_consumer'] = data_deep['responseto_label'].cat.codes
data_deep['timely_response'] = data_deep['timely_response_label'].cat.codes
data_deep['consumer_disputed'] = data_deep['disputed_label'].cat.codes

# Split into train and test - stratify over Issue
data_train, data_test = train_test_split(data_deep, test_size = 0.3,random_state=42)
```

## Loading BERT Pretrained

```python
# Name of the BERT model to use
Load_bert = LoadBert
bert_model = Load_bert.loadModel(modelName='bert-base-uncased')


# Load transformers config and set output_hidden_states to False
config = Load_bert.config(Bert_Configuration= BertConfig,
                          modelName= bert_model,
                          outputhidden_states= False)


# Load BERT tokenizer
tokenizer = Load_bert.tokenizer(tokenizer=BertTokenizerFast,
                                modelName=bert_model,
                                config= config)

# Load the Transformers BERT model
transformer_model = Load_bert.transformerModel(BertClass= TFBertModel,
                                               modelName= bert_model,
                                               config= config)

# Load the MainLayer
bert = transformer_model.layers[0]


# Max length of tokens
max_length = 70

# Build your model input
input_ids = Input(shape=(max_length,), name='input_ids', dtype='int32')
inputs = {'input_ids': input_ids}

# Load the Transformers BERT model as a layer in a Keras model
bert_model = bert(inputs)[1]

# Setting dropouts to tune the model 
dropout = Dropout(config.hidden_dropout_prob, name='pooled_output')

pooled_output = dropout(bert_model, training=False)

attention_dropout = Dropout(config.attention_probs_dropout_prob, name='attention_probs_dropout')

bert_model = attention_dropout(bert_model, training=True)

config.hidden_dropout_prob = 0.2

config.attention_probs_dropout_prob = 0.2
```

```python
# build your model output
product = Dense(units=len(data_train.product_label.value_counts()),
                activation = 'sigmoid', 
                kernel_initializer=GlorotNormal(),
                kernel_regularizer=regularizers.L1L2(l1=1e-4, l2=1e-3),
                       name='product')(pooled_output)

company_response = Dense(units=len(data_train.responseto_label.value_counts()),
                         activation = 'sigmoid', 
                         kernel_initializer=GlorotNormal(), 
                         kernel_regularizer=regularizers.L1L2(l1=1e-4, l2=1e-3),
                       name='response')(pooled_output)

timely_response = Dense(units = len(data_train.timely_response_label.value_counts()),
                        activation = 'sigmoid',
                        kernel_initializer = GlorotNormal(),
                        kernel_regularizer = regularizers.L1L2(l1=1e-4, l2=1e-3),
                        name ='timely_response')(pooled_output)

consumer_disputed= Dense(units = len(data_train.disputed_label.value_counts()),
                         activation = 'sigmoid', 
                         kernel_initializer =GlorotNormal(), 
                         kernel_regularizer =regularizers.L1L2(l1=1e-4, l2=1e-3),
                         name ='disputed')(pooled_output)

outputs = {'product': product,
           'response': company_response,
           'timely_response':timely_response,
           'disputed': consumer_disputed
          }
baseline_model = Model(inputs = inputs,
                       outputs = outputs,
                       name = 'BERT_MultiLabel_uncased'
                      )
# Take a look at the model
baseline_model.summary()
```

```python
# Set loss and metrics for the model
loss = {'product':CategoricalCrossentropy(from_logits = True),
        'response':CategoricalCrossentropy(from_logits = True),
        'timely_response':BinaryCrossentropy(from_logits = True),
        'disputed':BinaryCrossentropy(from_logits = True)}

optimizer = SGD(learning_rate = 0.00001,momentum = 0.99)

metric = [tf.keras.metrics.AUC(),
          tf.keras.metrics.Recall(),
          tf.keras.metrics.Precision(),
          tf.keras.metrics.CategoricalAccuracy(),
          tf.keras.metrics.FalseNegatives(),
          tf.keras.metrics.FalsePositives(),
          tf.keras.metrics.TrueNegatives(),
          tf.keras.metrics.TruePositives()
         ]
```

```python
# Reading y_train output from data_train for the model
y_product_train = loadData.get_target(data_train, 'product')
y_response_train = loadData.get_target(data_train,'company_response_to_consumer')
y_disputed_train = loadData.get_target(data_train, 'consumer_disputed')
y_timely_response_train = loadData.get_target(data_train, 'timely_response')


# Reading y_test output from data_test for model validation
y_product_test = loadData.get_target(data_test, 'product')
y_response_test = loadData.get_target(data_test,'company_response_to_consumer')
y_disputed_test = loadData.get_target(data_test, 'consumer_disputed')
y_timely_response_test = loadData.get_target(data_test, 'timely_response')

# consolidating for ease of legibility
y_train ={
    'product': y_product_train,
    'response': y_response_train,
    'timely_response':y_timely_response_train,
    'disputed': y_disputed_train}

# consolidating for ease of legibility
y_test ={
    'product': y_product_test,
    'response': y_response_test,
    'timely_response':y_timely_response_test,
    'disputed': y_disputed_test
         }
```

```python
# Tokenize the x_train to get train_tokens(Takes A While.)
train_tokens = loadData.get_tokens(tokenizer=tokenizer,
                                   df= data_train,
                                   text_col='clean_narrative',
                                   max_len= max_length)
x_train = {'input_ids': train_tokens['input_ids']}
```
